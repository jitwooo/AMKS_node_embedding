```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from running import rw_directory
```

```python
g_size = 6

g = nx.cycle_graph(g_size)
pos = nx.layout.circular_layout(g)
plt.grid(False)
nx.draw_networkx(g, pos, node_color='navy',  with_labels=False,font_size=10)

plt.savefig(rw_directory.result_path(1, 'ring_graph.pdf'), bbox_inches='tight', pad_inches=0)
```

```python
# 计算图拉普拉斯矩阵特征值
eigenvalues, eigenvectors = np.linalg.eigh(nx.laplacian_matrix(g, nodelist=sorted(g)).todense())
eigenvectors = eigenvectors.T
eigenvectors
# eigenvectors[index]是两层list，，，
```

```python
fig, axis = plt.subplots(nrows=len(eigenvalues), figsize=(g_size, eigenvalues.shape[0]))
for index in range(g_size):
    axis[index].scatter(range(g_size), eigenvectors[index].tolist()[0], s=10, color='black')
    axis[index].plot(range(g_size), eigenvectors[index].tolist()[0], color='black')
    # axis[index].set_title(r"$\lambda$ = {:2f}".format(eigenvalues[index]))
    axis[index].set_title(r"$\lambda$" + f" = {eigenvalues[index]:.2f}")
    axis[index].set_ylim(bottom=-1.0, top=1.0)
    # axis[index].set_xlabel("Element index")
    # axis[index].set_ylabel("Element value")
plt.subplots_adjust(hspace=1.5)
plt.savefig(rw_directory.result_path(1, 'eigenvectors.pdf'), bbox_inches='tight', pad_inches=0)


# fig, axis = plt.subplots(nrows=len(eigenvalues), figsize=(g_size, eigenvalues.shape[0]))
# for index in range(g_size):
#     # plt.figure(figsize=(g_size, eigenvalues.shape[0]))
#     plt.scatter(range(g_size), eigenvectors[index].tolist()[0], s=10, color='black')
#     plt.plot(range(g_size), eigenvectors[index].tolist()[0], color='black')
#     # axis[index].set_title(r"$\lambda$ = {:2f}".format(eigenvalues[index]))
#     plt.title(r"$\lambda$" + f" = {eigenvalues[index]:.2f}")
#     plt.ylim(bottom=-1.0, top=1.0)
#     plt.xlabel("Element index")
#     plt.xlabel("Element value")
#     plt.savefig(rw_directory.result_path(1, f'eigenvectors{index}.pdf'), bbox_inches='tight', pad_inches=0)
#     plt.show()

```
